# Semantic analysis with graph transformations

```python
def accuracy(file_name):
    with open(file_name, 'r') as commonsense:
        dev_acc = []
        train_acc = []
        line = commonsense.readline()
        while line is not None and line != "":
            if line.startswith("Dev accuracy"):
                dev_acc.append(float(line.split(": ")[-1].strip()))
            if line.startswith("Train accuracy"):
                train_acc.append(float(line.split(": ")[-1].strip()))
            line = commonsense.readline()
        epoch = [x for x in range(0, len(dev_acc))]
        train_acc.insert(0, dev_acc[0])
        from matplotlib import pyplot as plt
        fig = plt.figure(figsize=(10, 5), dpi=100)
        plt.plot(epoch, dev_acc, label="Dev accuracy")
        plt.plot(epoch, train_acc, label="Train accuracy")
        plt.legend()
        
        plt.title('Accuracy over dev and train data')
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        print("Max dev accuracy: {0} reached in the {1}th epoch".format(max(dev_acc), dev_acc.index(max(dev_acc))))
        print("Max train accuracy: {0} reached in the {1}th epoch".format(max(train_acc), train_acc.index(max(train_acc))))
        print("Last dev accuracy: {0}".format(dev_acc[-1]))
        print("Average dev accuracy after ten epochs: {0}".format(sum(dev_acc[9:])/len(dev_acc[9:])))
```
